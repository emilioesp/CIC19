```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
import numpy as np 
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
string = '201116' #fecha de base de datos
path2 = 'C:/Users/Emilio EG/Documents/ITAM/CIC19/Reportes/'+string+'/'

hoy=datetime.datetime.today()
corte = hoy - datetime.timedelta(days=14) #hoy-12dias

base_covid = pd.read_csv(path2+string+'COVID19MEXICO'+'.csv',encoding='latin1') 
base_covid
```

```python
base_covid=base_covid[base_covid['RESULTADO_LAB']==1] #positivo covid
base_covid
```

```python
base_covid.loc[base_covid['FECHA_DEF'] == '9999-99-99', 'FECHA_DEF'] = pd.NaT #9999-99-99 pasa a ser NaT
base_covid['FECHA_DEF']=pd.to_datetime(base_covid['FECHA_DEF']) #los valores de la columna pasan a ser datetime
base_covid['FECHA_SINTOMAS']=pd.to_datetime(base_covid['FECHA_SINTOMAS']) #los valores de la columna pasan a ser datetime

base_def=base_covid.dropna() #eliminamos los NA, NaN y NaT
base_def
```

```python
estados=pd.read_excel('C:/Users/Emilio EG/Documents/ITAM/CIC19/datos/generales/Estados.xlsx')
estados
```

```python
estados.loc[8]
```

```python
defs={}
defs_tot={}
defs_avg={}

for i in range(len(estados)):
    defs[i]=base_def[base_def['ENTIDAD_UM']==i+1].pivot_table(index=['FECHA_DEF'], aggfunc='size') #Contar muertes por dia en cada municipio
    defs[i]=pd.DataFrame(defs[i]) #lo anterior a dataframe
    idx=pd.date_range(defs[i].index[0],hoy) #creamos un indice de fechas desde el primer registro hasta hoy
    defs[i]=defs[i].reindex(idx, fill_value=0) #reindexamos el df para no saltarnos fechas
    defs_tot[i]=defs[i].cumsum()*(100000/estados['POBLACION'][i]) #el acumulado dia a dia por 100 mil habs
    defs_avg[i]=defs[i].rolling(window=7).mean()*(100000/estados['POBLACION'][i]) #defunciones diarias por 100 mil habs (promedio movil)
    defs_avg[i]=defs_avg[i].dropna()
```

```python
con={}
con_tot={}
con_avg={}

for i in range(len(estados)):
    con[i]=base_covid[base_covid['MUNICIPIO_RES']==i+1].pivot_table(index=['FECHA_SINTOMAS'], aggfunc='size') #contar casos nuevos por dia
    con[i]=pd.DataFrame(con[i]) #lo anterior como dataframe
    idx=pd.date_range(con[i].index[0],hoy) #creamos un indice de fechas desde el primer registro hasta hoy
    con[i]=con[i].reindex(idx, fill_value=0) #reindexamos el df para no saltarnos fechas
    con_tot[i]=con[i].cumsum()*(100000/estados['POBLACION'][i]) #el acumulado dia a dia por 100 mil habs
    con_avg[i]=con[i].rolling(window=7).mean()*(100000/estados['POBLACION'][i]) #promedio movil de casos diarios por 100 mil habs
    con_avg[i]=con_avg[i].dropna()
```

```python
defus=pd.DataFrame(np.zeros(32))
conts=pd.DataFrame(np.zeros(32))

for i in range(len(estados)):
    defus.loc[i]=defs[i].sum()[0]*(100000/estados['POBLACION'][i])
    conts.loc[i]=con[i].sum()[0]*(100000/estados['POBLACION'][i])
```

```python
plt.scatter(conts[0],defus[0])
plt.title('Contagios y defunciones acumulados por 100 mil habitantes\n en cada estado')
plt.xlabel('Contagios')
plt.ylabel('Defunciones')

for i, txt in enumerate(estados['ABREVIATURA']):
    plt.annotate(txt, (conts.loc[i], defus.loc[i]))
    
plt.savefig('Estados_totales.png', bbox_inches='tight')
```

```python
tot_avg={}

for i in range(len(estados)):
    defs_avg[i] = defs_avg[i].reindex(con_avg[i].index, fill_value=0)
    tot_avg[i]=defs_avg[i].merge(con_avg[i],left_index=True,right_index=True)
    tot_avg[i]=tot_avg[i].rename(columns={'0_x':'DEF','0_y':'CON'})
    tot_avg[i]=tot_avg[i].loc[tot_avg[i].index[0:-14]]
    tot_avg[i]= tot_avg[i].dropna() 
    plt.plot(tot_avg[i]['CON'],tot_avg[i]['DEF'],'.-')
    plt.title(estados['ENTIDAD_FEDERATIVA'][i])
    plt.xlabel('Contagios')
    plt.ylabel('Defunciones')
    plt.annotate('0', (tot_avg[i]['CON'][0], tot_avg[i]['DEF'][0]))
    plt.annotate('1', (tot_avg[i]['CON'][-1], tot_avg[i]['DEF'][-1]))
    #plt.savefig(estados['ENTIDAD_FEDERATIVA'][i]+ 'x100k.png', bbox_inches='tight')  
    plt.show()
```

```python
def_nac=base_def.pivot_table(index=['FECHA_DEF'], aggfunc='size')
def_nac=pd.DataFrame(def_nac)
def_nac=def_nac.rolling(window=7).mean()*(100000/119530753)
def_nac=def_nac.dropna()

con_nac=base_covid.pivot_table(index=['FECHA_SINTOMAS'], aggfunc='size')
con_nac=pd.DataFrame(con_nac)
con_nac=con_nac.rolling(window=7).mean()*(100000/119530753)
con_nac=con_nac.dropna()

def_nac=def_nac.reindex(con_nac.index,fill_value=0)
nacional=def_nac.merge(con_nac,left_index=True,right_index=True)
nacional=nacional.rename(columns={'0_x':'DEF','0_y':'CON'})
nacional=nacional.loc[nacional.index[0:-14]]
#nacional=nacional.dropna()

plt.plot(nacional['CON'],nacional['DEF'],'.-')
plt.title('NACIONAL')
plt.xlabel('Contagios')
plt.ylabel('Defunciones')
plt.annotate('0', (nacional['CON'][0], nacional['DEF'][0]))
plt.annotate('1', (nacional['CON'][-1], nacional['DEF'][-1]))
#plt.savefig('Nacionalx100k.png', bbox_inches='tight')  
plt.show()
```
